```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler,OneHotEncoder
import pickle
```

```python
# Data 
df=pd.read_csv("C:\\Users\\DELL\\Downloads\\Churn_Modelling.csv")
```

```python
df
```

```python
# Data Preprocessing
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df
```

```python
label=LabelEncoder()
df['Gender']=label.fit_transform(df['Gender'])
```

```python
df
```

```python
one_hot=OneHotEncoder()
geo_encoder=one_hot.fit_transform(df[['Geography']])
geo_encoder
```

```python
one_hot.get_feature_names_out(['Geography'])
```

```python
geo_df=pd.DataFrame(geo_encoder.toarray(),columns=one_hot.get_feature_names_out(['Geography']))
```

```python
geo_df
```

```python
import glob
df=pd.concat([df.drop('Geography',axis=1),geo_df],axis=1)
```

```python
df
```

```python
# Save the encoders as pickle file
with open('gender_encoders.pkl', 'wb') as f:
    pickle.dump(label, f)

with open('geo_enoders.pkl','wb') as f:
    pickle.dump(one_hot,f)
```

```python
# Train and Test
X=df.drop('Exited',axis=1)
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

# Scaling features
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Saving the pickle file
import pickle
with open('scaler_df.pkl','wb') as file:
    pickle.dump(scaler, file)
```

```python
## Building ANN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
## Building ANN Model

from keras.layers import Dense
from keras.models import Sequential
model=Sequential(
    [
        Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
        Dense(128,activation='relu'),
        Dense(64,activation='relu'),
        Dense(1,activation='sigmoid')
    ]
)
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# Compiling the model
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Setting up Tensorboard
import datetime
import tensorflow as tf
log_dir="logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# Setting up Early Stopping
early_stopping_callback=tf.keras.callbacks.EarlyStopping(patience=10,monitor='val_loss',restore_best_weights=True)
```

```python
# Training the model
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),
                  epochs=100,callbacks=[tensorflow_callback,early_stopping_callback])
```

```python
# Saving the model
model.save('model.h5')
```
